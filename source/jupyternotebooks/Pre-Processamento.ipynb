```python
from collections import defaultdict

# Agrupar as horas dos eventos por ano
events_by_month = defaultdict(list)
events_by_year = defaultdict(list)

for date in events_datetime:
    events_by_month[date.month].append(date.hour)
    events_by_year[date.year].append(date.hour)
```

```python
# Lista para armazenar as horas de eventos de cada ano e os respectivos rótulos (anos)
hours_list = []
years_labels = []

for year in sorted(events_by_year.keys()):
    hours_list.append(events_by_year[year])
    years_labels.append(year)
```

```python
# Plotando os boxplots Hours by Years
plt.figure(figsize=(12, 8))
plt.boxplot(hours_list, patch_artist=True)
plt.suptitle('MOHO-IAG(USP) / Natural Labeled Events')
plt.title('Distribuição dos Eventos por Horas do Dia para Cada Ano')
plt.xlabel('Ano')
plt.ylabel('Hora do Dia')
plt.xticks(range(1, len(years_labels) + 1), years_labels)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
